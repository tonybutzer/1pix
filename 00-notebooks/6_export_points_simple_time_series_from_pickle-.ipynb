```python
import panel as pn
import numpy as np
import xarray as xr
import holoviews as hv
import hvplot.xarray
import pandas as pd
import param

pn.extension()

save_x = 0
save_y = 0

xy_df = pd.DataFrame() 
df_pane = pn.pane.DataFrame(xy_df)

#ds = xr.tutorial.open_dataset('air_temperature')
button = pn.widgets.Button(name='Pickle Points', button_type='primary')
button

image = ds['etasw'][0].hvplot('x', 'y', cmap='viridis', rasterize=True, dynamic=False)
stream = hv.streams.Tap(source=image, x=0, y=0)

@pn.depends(stream.param.x, stream.param.y)
#@pn.depends(stream.param.x)
def timeseries(x, y):
    global save_x
    global save_y
    global xy_df
    global df_pane
    global bpcnt

    if (x != save_x):
        save_x = x
        save_y = y
        new_row = {'x':save_x, 'y':save_y}
        xy_df = xy_df.append(new_row, ignore_index=True)
        df_pane = pn.pane.DataFrame(xy_df)

    text.value = 'Clicked {0} times b{1} {2}  {3}'.format(button.clicks, bpcnt, save_x, save_y)
    return pn.Row(pn.panel(ds.sel(x=x, y=y, method='nearest').hvplot('year_month'), width=350, height=200), df_pane)

    
text = pn.widgets.TextInput(value='Ready')
df_pane = pn.pane.DataFrame(xy_df)

bpcnt=0
def b(event):
    global xy_df
    new_row = {'x':save_x, 'y':save_y}
    xy_df = xy_df.append(new_row, ignore_index=True)
    #row = {'x':savenew__x, 'y':save_y}
    #xy_df = xy_df.append(new_row, ignore_index=True)
    #df_pane = pn.pane.DataFrame(xy_df)
    #df_pane.force_new_dynamic_value(xy_df)
    text.value = 'Pickle Clicked {0} times {1}  {2}  {3]}'.format(button.clicks, save_x, save_y, xy_df)
    do_pickle(xy_df)
    bpcnt = bpcnt+1
    
    
button.on_click(b)

button1 = pn.widgets.Button(name='Click me', button_type='primary')
#button

#The clicks parameter will report the number of times the button has been pressed:

button1.clicks

#The on_click method can trigger function when button is clicked:

text1 = pn.widgets.TextInput(value='Ready')

def b1(event):
    text1.value = 'Clicked {0} times'.format(button.clicks)
    
button.on_click(b1)

stock_commands = ['NO-OP','SAVE POINTS', 'CLEAR POINT']
collection_select2 = pn.widgets.Select(name='commands', options=stock_commands)

# pn.Row(   
#     pn.Column(pn.Row(text, button, df_pane), pn.panel(image,width=400,height=300), 
#               pn.panel(timeseries, width=300, height=200))
# )

@pn.depends(collection_select2.param.value)
def do_command(commands):
    do_pickle(xy_df)
    return("HI")
    
pn.Row(   
    pn.Column(pn.Row(text, button), pn.panel(image,width=400,height=300), 
              pn.panel(timeseries, width=300, height=200), collection_select2),
)

#xy_df
```
